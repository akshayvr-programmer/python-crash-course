Section A: 15.  Loops: For loops and While loops
A loop in any programming language is a way of repeatedly executing a line of code.

Now, we'll be talking about while loops:
While loops are a way of telling python that while a certain condition is true or false, execute the lines of code: 

We'll be understanding a while loop using a sequence known as the fibonacci sequence. In a fibonacci sequence, every term is a sum of the other two terms.

Example of Fibonacci Sequence = 0, 1, 1, 2, 3, 5, 8, 13......

We'll have to specify until how many terms we want the sequence to be printed.

Let's say that I want the first 6 terms of the sequence. Let the terms be 6.
the first number is always 0 and the second number is always 1. The third number will be the sum of the first and second number. We'll have to continuously increase the value of the first and second numbers until the value of the nth term is reached. Lets try writing some code for it:

```
def fib(n):
  nth_terms = int(input("Enter the value of the n'th term"))
  n_1 = 0
  n_2 = 1
  counter = 0
  while (counter < nth_terms):
    print(n_1)
    nth = n_1 + n_2
    n_1 = n_2
    n_2 = nth

    counter += 1
    if counter == nth_terms:
      break

fib(0)
```
